```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
import statsmodels.api as sm
```

```python
from statsmodels.tsa.stattools import acovf, acf, pacf, pacf_yw, pacf_ols
```

```python
# NON STATIONARY
df1 = pd.read_csv('./Data Files/airline_passengers.csv', index_col='Month', parse_dates=True)
df1.index.freq = 'MS'
```

```python
# STATIONARY
df2 = pd.read_csv('./Data Files/DailyTotalFemaleBirths.csv', index_col='Date', parse_dates=True)
df2.index.freq = 'D'
```

```python
df = pd.DataFrame({'a':[13,5,11,12,9]})
```

```python
df
```

```python
acf(df['a'])
```

```python
# Yule Walker Equation
pacf_yw(df['a'], nlags=4, method='mle') #or unbiased method
```

```python
# Ordinary least squares
pacf_ols(df['a'], nlags=4)
```
