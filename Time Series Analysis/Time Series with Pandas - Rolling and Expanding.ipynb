```python
import pandas as pd
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('./Data Files/starbucks.csv', index_col='Date', parse_dates=True)
```

```python
df['Close'].plot(figsize=(12,5))
```

```python
df['Close'].plot(figsize=(12,5))
df.rolling(window=7).mean()['Close'].plot(figsize=(20,5))
```

```python
df['Close: 30 Day Mean'] = df['Close'].rolling(window=30).mean()
```

```python
df[['Close','Close: 30 Day Mean']].plot(figsize=(12,3))
```

```python
df['Close'].expanding().mean().plot(figsize=(12,5))
```
